```python
import pandas as pd
```

```python
# Load the hourly-based data
df = pd.read_csv('hour.csv', header=0)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# The columns of the explanatory and target variables
cols = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']

# The index of the target variable
target_idx = len(cols) - 1

# For each index of the explanatory variable
for exp_var_idx in range(len(cols) - 1):
    # Plot data and regression model comprised of the target variable and the current explanatory variable  
    sns.lmplot(x = cols[exp_var_idx], y = cols[target_idx], data = df, line_kws={'color': 'red'})
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Get Pearson product-moment correlation coefficients
cm = np.corrcoef(df[cols].values.T)

# Apply np.argmin / np.argmax to the correlation matrix (cm) and identify the explanatory variables that have the min / max correlation 
# with the target (cnt)
max_neg_cor_exp_var = cols[np.argmin(cm[range(len(cols) - 1), len(cols) - 1])]
max_pos_cor_exp_var = cols[np.argmax(cm[range(len(cols) - 1), len(cols) - 1])]

# Get heat map
plt.figure(figsize=(8,8))
hm = sns.heatmap(cm,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 15},
                 yticklabels=cols,
                 xticklabels=cols)

# Show heat map
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Get the target values
y = df['cnt'].values.reshape(-1,1)

# Standardize the target values
sc_y = StandardScaler()
y_std = sc_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Get the Mean Squared Error. Here:
# explanatory_var is an explanatory variable
# y_std is the standardized target values
def get_MSE(explanatory_var, y_std):
    # Standardize explanatory variable values 
    sc_X = StandardScaler()
    X = df[[explanatory_var]].values
    X_std = sc_X.fit_transform(X)
    
    # Randomly choose 70% of the data for training
    X_train, X_test, y_train, y_test = train_test_split(X_std, y_std, test_size = 0.3, random_state = 0)

    # Train the model on the training set
    slr = LinearRegression()
    slr.fit(X_train, y_train)
    
    # Test the model on the testing set
    y_test_pred = slr.predict(X_test)    

    # Plot the model on the testing set
    plt.scatter(X_test, y_test, c = 'steelblue', edgecolor = 'white', s = 70)
    plt.plot(X_test, y_test_pred, color = 'black', lw = 2)  
    plt.xlabel(explanatory_var)
    plt.ylabel('cnt')
    plt.show()
    
    # Print Slope and Intercept
    print('Slope: %.3f' % slr.coef_[0])
    print('Intercept: %.3f' % slr.intercept_)
    
    # Get the mean squared error
    mean_squared_error_val = mean_squared_error(y_test, y_test_pred)
    # Print the mean squared error 
    print('Mean squared error: %.3f' % mean_squared_error_val)
    
    # Return the mean squared error
    return mean_squared_error_val
```

```python
# Get mean squared error of the two models
# Hint: call the function that we defined above
MSE_max_neg_cor = get_MSE(max_neg_cor_exp_var, y_std)
MSE_max_pos_cor = get_MSE(max_pos_cor_exp_var, y_std)

# Print empty line
print()

# Print the name of the explanatory variable that results in a model with the lower mean squared error
if MSE_max_neg_cor < MSE_max_pos_cor:
    print('The explanatory variable that results in a model with the lower mean squared error is: ' + max_neg_cor_exp_var)
elif MSE_max_neg_cor > MSE_max_pos_cor:
    print('The explanatory variable that results in a model with the lower mean squared error is: ' + max_pos_cor_exp_var)
else:
    print('The explanatory variable that results in a model with the lower mean squared error is: ' + max_neg_cor_exp_var + ' and ' + max_pos_cor_exp_var)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = log(x)
x = np.arange(10**(-6), 10**6, 1)
y = np.log(x)

plt.xlabel('x', fontsize = 20)
plt.ylabel('log(x)', fontsize = 20)

plt.xticks([0, 500000, 1000000], fontsize=20)
# plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.yticks(fontsize=20)

plt.tight_layout()
plt.plot(x, y, 'r--')
plt.savefig(fname = 'log.pdf', format='pdf', dpi=300)
plt.show()
```

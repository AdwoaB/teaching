# Data Preprocessing

In this case study, we will work on the [House Prices (Advanced Regression Techniques) dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques).

```python
import pandas as pd

# Load the training data
df_train = pd.read_csv('./data/house_prices_advanced_regression_techniques/train.csv',
                       header=0)

# Load the testing data
df_test = pd.read_csv('./data/house_prices_advanced_regression_techniques/test.csv',
                      header=0)

# Get the name of the target
target = 'SalePrice'

# Print the dimension of df_train and df_test
print(df_train.shape)
print(df_test.shape)
```

## Handling the Identifiers

### Combining the training and testing data
The code below shows how to combine the training and testing data (using pandas concat).

```python
# Combine df_train and df_test
df = pd.concat([df_train, df_test], sort=False)
```

### Identifying the identifiers
The code below shows how to find identifiers from the data.

```python
def id_checker(df):
    """
    The identifier checker

    Parameters
    ----------
    df : dataframe
    
    Returns
    ----------
    The dataframe of identifiers
    """
    
    # Get the identifiers
    df_id = df[[var for var in df.columns 
                if df[var].nunique(dropna=True) == df[var].notnull().sum()]]
                
    return df_id
```

```python
# Call id_checker on df
df_id = id_checker(df)

# Print the first 5 rows of df_id
df_id.head()
```

### Removing the Identifiers
The code below shows how to remove the identifiers from data (using pandas DataFrame.drop).

```python
# Remove the identifiers from df_train
df_train = df_train.drop(columns=df_id.columns)

# Remove the identifiers from df_test
df_test = df_test.drop(columns=df_id.columns)

# Print the dimension of df_train and df_test
print(df_train.shape)
print(df_test.shape)
```

```python
# Print the first 10 rows of df_train
df_train.head(10)
```

## Handling Missing Data

```python
# Print the first 10 rows of df_train
df_train.head(10)
```

### Combining the training and testing data
The code below shows how to combine the training and testing data (using pandas concat).

```python
# Combine df_train and df_test
df = pd.concat([df_train, df_test], sort=False)
```

### Identifying missing values

The code below shows how to find variables with NaN (using pandas DataFrame.isna), their proportion of NaN and dtype (data type objects, using pandas Series.dtype).

```python
def nan_checker(df):
    """
    The NaN checker

    Parameters
    ----------
    df : dataframe
    
    Returns
    ----------
    The dataframe of variables with NaN, their proportion of NaN and dtype
    """
    
    # Get the variables with NaN, their proportion of NaN and dtype
    df_nan = pd.DataFrame([[var, df[var].isna().sum() / df.shape[0], df[var].dtype]
                           for var in df.columns if df[var].isna().sum() > 0],
                          columns=['var', 'proportion', 'dtype'])
    
    # Sort df_nan in accending order of the proportion of NaN
    df_nan = df_nan.sort_values(by='proportion', ascending=False)
    
    return df_nan
```

```python
# Call nan_checker on df
df_nan = nan_checker(df)

# Print df_nan
df_nan.reset_index(drop=True)
```

```python
# Print the unique dtype of the variables with NaN
pd.DataFrame(df_nan['dtype'].unique(), columns=['dtype'])
```

The code below shows how to use dtype to select variables with missing values in the combined training and testing data.

```python
# Get the variables with missing values, their proportion of missing values and dtype
df_miss = df_nan[df_nan['dtype'] == 'float64']

# Print df_miss
df_miss.reset_index(drop=True)
```

### Removing missing values
The code below shows how to remove rows with missing values (here we use numpy intersect1d to find the intersection of two series).

```python
import numpy as np

# Remove rows with missing values from df_train
df_remove_train = df_train.dropna(subset=np.intersect1d(df_miss['var'], 
                                                        df_train.columns),
                                 inplace=False)

# Remove rows with missing values from df_test
df_remove_test = df_test.dropna(subset=np.intersect1d(df_miss['var'], 
                                                      df_test.columns),
                               inplace=False)

# Print the dimension of the training and testing data
print(df_remove_train.shape)
print(df_remove_test.shape)
```

### Imputing missing values
The code below shows how to use the mean to a numerical variable to impute its missing values (using sklearn SimpleImputer).

```python
from sklearn.impute import SimpleImputer

# The SimpleImputer
si = SimpleImputer(missing_values=np.nan, strategy='mean')

# Make a copy of df_train and df_test
df_impute_train = df.iloc[:df_train.shape[0],:].copy(deep=True)
df_impute_test = df.iloc[df_train.shape[0]:,:].copy(deep=True)

# Impute the variables with missing values in df_impute_train and df_impute_test 
df_impute_train[df_miss['var']] = si.fit_transform(df_impute_train[df_miss['var']])
df_impute_test[df_miss['var']] = si.transform(df_impute_test[df_miss['var']])

# Print the first 10 rows of df_impute_train
df_impute_train.head(10)
```

## Encoding the data

### Combining the training and testing data
The code below shows how to combine the training and testing data (using pandas concat).

```python
# Combine df_impute_train and df_impute_test
df = pd.concat([df_impute_train, df_impute_test], sort=False)

# Print the unique dtype of variables in df
pd.DataFrame(df.dtypes.unique(), columns=['dtype'])
```

### Identifying the Categorical Variables

The code below shows how to find the categorical variables that have object as dtype (using pandas.Series.dtype).

```python
def cat_var_checker(df):
    """
    The categorical variable checker

    Parameters
    ----------
    df: the dataframe
    
    Returns
    ----------
    The dataframe of categorical variables and their number of unique value
    """
    
    # Get the dataframe of categorical variables and their number of unique value
    df_cat = pd.DataFrame([[var, df[var].nunique(dropna=False)]
                           for var in df.columns if df[var].dtype == 'object'],
                          columns=['var', 'nunique'])
    
    # Sort df_cat in accending order of the number of unique value
    df_cat = df_cat.sort_values(by='nunique', ascending=False)
    
    return df_cat
```

```python
# Call cat_var_checker on df
df_cat = cat_var_checker(df)

# Print the dataframe
df_cat.reset_index(drop=True)
```

### Encoding the features
The code below shows how to encode the categorical features in the combined training and testing data (using pandas.get\_dummies).

```python
# One-hot-encode the categorical features in the combined training and testing data
df = pd.get_dummies(df, columns=df_cat['var'])

# Print the first 10 rows (samples) of df
df.head(10)
```

### Separating the training and testing data
The code below shows how to separate the training and testing data.

```python
# Separating the training data
df_train = df.iloc[:df_train.shape[0], :]

# Separating the testing data
df_test = df.iloc[df_train.shape[0]:, :]

# Print the dimension of the training and testing data
print(df_train.shape)
print(df_test.shape)
```

## Scaling the data

### Standardization
The code below shows how to standardize the data (using sklearn StandardScaler).

```python
from sklearn.preprocessing import StandardScaler

# The StandardScaler
ss = StandardScaler()

# Standardize the training data
df_ss_train = pd.DataFrame(ss.fit_transform(df_train), columns=df_train.columns)

# Standardize the testing data
df_ss_test = pd.DataFrame(ss.transform(df_test), columns=df_test.columns)
```

### Min-Max Normalization
The code below shows how to normalize the data (using sklearn MinMaxScaler). 

```python
from sklearn.preprocessing import MinMaxScaler

# The MinMaxScaler
mms = MinMaxScaler()

# Normalize the training data
df_mms_train = pd.DataFrame(mms.fit_transform(df_train), columns=df_train.columns)

# Normalize tthe testing data
df_mms_test = pd.DataFrame(mms.transform(df_test), columns=df_test.columns)
```

### Standardization VS Min-Max Normalization
The code below shows the difference in the range of the variables in the training data transformed by StandardScaler and MinMaxScaler

```python
# Get the min and max of each feature in the training data after standardization
mins_ss_train, maxs_ss_train = df_ss_train.min(axis=1), df_ss_train.max(axis=1)

# Get the min and max of each feature in the training data after min-max normalization
mins_mms_train, maxs_mms_train = df_mms_train.min(axis=1), df_mms_train.max(axis=1)

# Get the dataframe of min and max
df_min_max_train = pd.DataFrame([[mins_ss_train.min(), 
                                  mins_ss_train.max(), 
                                  maxs_ss_train.min(), 
                                  maxs_ss_train.max()],
                                 [mins_mms_train.min(), 
                                  mins_mms_train.max(), 
                                  maxs_mms_train.min(), 
                                  maxs_mms_train.max()]],
                                index=['StandardScaler', 'MinMaxScaler'],
                                columns=['min of the mins', 
                                         'max of the mins',
                                         'min of the maxs',
                                         'max of the maxs'])

# Print df_min_max_train
round(df_min_max_train, 3)
```

The code below shows the difference in the robustness to outliers between StandardScaler and MinMaxScaler.

```python
# Copy df_train
df_outlier_train = df_train.copy(deep=True)

# Add the outlier to the first sample of the first variable in df_outlier_train
df_outlier_train.iloc[0, 0] = 10 ** 6

# Standardize df_outlier_train
df_outlier_ss_train = ss.fit_transform(df_outlier_train)

# Normalize df_outlier_train
df_outlier_mms_train = mms.fit_transform(df_outlier_train)

# Get the dataframe of the first variablefeature in df_train, df_outlier_train, df_outlier_ss_train and df_outlier_mms_train
df_outlier_ss_mms_train = pd.DataFrame(np.hstack((df_train.iloc[:, 0].values.reshape(-1, 1),
                                                  df_outlier_train.iloc[:, 0].values.reshape(-1, 1),
                                                  df_outlier_ss_train[:, 0].reshape(-1, 1),
                                                  df_outlier_mms_train[:, 0].reshape(-1, 1))),
                                       columns=['df_train',
                                                'df_outlier_train', 
                                                'df_outlier_StandardScaler_train', 
                                                'df_outlier_MinMaxScaler_train']).astype({'df_train':int,
                                                                                          'df_outlier_train':int,
                                                                                          'df_outlier_StandardScaler_train':float,
                                                                                          'df_outlier_MinMaxScaler_train':float})

# Print the first 10 rows of df_outlier_ss_mms_train
round(df_outlier_ss_mms_train.head(10), 3)
```

### The Normal Equation

The code below shows how to implement the linear regression model using the normal equation.

```python
from sklearn.base import BaseEstimator, RegressorMixin

class LinearRegression_NE(BaseEstimator, RegressorMixin):
    """The linear regression model implemented using the normal equation"""

    def fit(self, X, y):
        """
        The fit function
        
        Parameters
        ----------
        X : the feature matrix
        y : the target vector
        """
        
        # The feature matrix with x0 (the dummy feature)
        Z = np.hstack((np.ones((X.shape[0], 1)), X))

        # Get the optimal solution using the normal equation
        self.w = np.linalg.inv(Z.T.dot(Z)).dot(Z.T).dot(y)
        
        # Get the net input (the predicted value of the target)
        net_input = self.net_input(Z)
                        
        # Get the cost (mean squared error)
        self.cost = ((y - net_input) ** 2).sum() / Z.shape[0]
        
    def net_input(self, Z):
        """
        Get the net input
        
        Parameters
        ----------
        Z : The feature matrix with x0 (the dummy feature)
        
        Returns
        ----------
        The net input (the predicted value of the target)
        """
        
        return Z.dot(self.w)

    def predict(self, X):
        """
        The predict function
        
        Parameters
        ----------
        X : the feature matrix
        
        Returns
        ----------
        The predicted value of the target
        """
        
        # Implement me
        return self.net_input(X)
```

```python
# The LinearRegression_NE
lr_ne = LinearRegression_NE()

# Train the simple linear regression model on the training data
lr_ne.fit(X_train[['LotFrontage']], y_train)

# Print the intercept (w0) and coefficient (w1) of the trained model
round(pd.DataFrame(lr_ne.w.reshape(1, -1), columns=['w0', 'w1']), 10)
```

### Batch Gradient Descent

The code below shows how to implement the linear regression model using batch gradient descent.

```python
class LinearRegression_BGD(BaseEstimator, RegressorMixin):
    """The linear regression model implemented using batch gradient descent"""
    
    def __init__(self, n_iter=100, eta=10 ** -2, random_state=0):
        # The number of iterations
        self.n_iter = n_iter
        
        # The learning rate
        self.eta = eta

        # The random state
        self.random_state = random_state

    def fit(self, X, y):
        """
        The fit function
        
        Parameters
        ----------
        X : the feature matrix
        y : the target vector
        """
        
        # The feature matrix with x0 (the dummy feature)
        Z = np.hstack((np.ones((X.shape[0], 1)), X))
        
        # The random number generator
        self.rgen = np.random.RandomState(seed=self.random_state)
        
        # Initialize the weight
        self.w = self.rgen.normal(loc=0.0, scale=0.01, size=Z.shape[1])
        
        # Initialize the cost
        self.cost = []

        # For each iteration
        for _ in range(self.n_iter):
            # Get the net_input (the predicted value of the target)
            net_input = self.net_input(Z)
            
            # Get the error
            error = y - net_input
                        
            # Get the mean squared error (mse)
            mse = (error ** 2).sum() / Z.shape[0]
            
            # Update the weight
            self.w += 2 * self.eta * Z.T.dot(error) / Z.shape[0]
                           
            # Update the cost
            self.cost.append(mse)

    def net_input(self, Z):
        """
        Get the net input
        
        Parameters
        ----------
        Z : The feature matrix with x0 (the dummy feature)
        
        Returns
        ----------
        The net input (the predicted value of the target)
        """
        
        return Z.dot(self.w)

    def predict(self, X):
        """
        The predict function
        
        Parameters
        ----------
        X : the feature matrix
        
        Returns
        ----------
        The predicted value of the target
        """
        
        # The feature matrix with x0 (the dummy feature)
        Z = np.hstack((np.ones((X.shape[0], 1)), X))
        
        return self.net_input(Z)
```

```python
# The LinearRegression_BGD
lr_bgd = LinearRegression_BGD()

# Train the simple linear regression model on the training data
lr_bgd.fit(X_train[['LotFrontage']], y_train)

# Print the intercept (w0) and coefficient (w1) of the trained model
round(pd.DataFrame(lr_bgd.w.reshape(1, -1), columns=['w0', 'w1']), 10)
```

```python
# Print the cost (mean squared error) in the last epoch
pd.DataFrame([lr_bgd.cost[-1]], columns=['mean squared error'])
```

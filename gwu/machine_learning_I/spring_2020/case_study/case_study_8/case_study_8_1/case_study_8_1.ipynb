# Data Preprocessing

## Loading data

In this case study, we will work on the [mnist dataset](https://www.tensorflow.org/datasets/catalog/mnist).

### Creating the directory for the data

```python
import os

# Make directory
directory = os.path.dirname(abspath + 'data/')
if not os.path.exists(directory):
    os.makedirs(directory)
```

### Downloading data to the directory

```python
import tensorflow_datasets as tfds

# Get the name of the data
data_name = "mnist"

# Load data
data = tfds.load(name=data_name,
                 split='train',
                 data_dir=abspath + 'data/',
                 as_supervised=True)
```

## Normalizing the data

```python
def normalize(data):
    """
    Normalize the data

    Parameters
    ----------
    data: the data
    label: the label
    
    Returns
    ----------
    The normalized data
    """

    # Normalize the data
    data_normalized = tf.cast(data, tf.float32) / 255

    return data_normalized
```

```python
def preprocess(data, label):
    """
    Preprocess the data using the pretrained model

    Parameters
    ----------
    data: the data
    label: the label
    
    Returns
    ----------
    The preprocessed data
    """

    # Normalize the data
    data_preprocessed = normalize(data)

    return data_preprocessed, label
```

```python
# Preprocess the data
data = data.map(preprocess)
```

## Shuffling, batching and prefetching the data

```python
# Shuffling the data
data = data.shuffle(buffer_size=1000, seed=42)

# Set the batch size
batch_size = 16

# Batch and prefetch the data
data = data.batch(batch_size).prefetch(1)
```

# Training

## Creating the directory of the saved model

```python
# Make directory
directory = os.path.dirname(abspath + 'model/')
if not os.path.exists(directory):
    os.makedirs(directory)
```

## Designing the model

```python
# Set the coding size
coding_size = 100

# Get the generator
generator = keras.models.Sequential([
    keras.layers.Dense(7 * 7 * 128, input_shape=[coding_size]),
    keras.layers.Reshape([7, 7, 128]),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2DTranspose(64, kernel_size=5, strides=2, padding='same',
                                 activation='selu'),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2DTranspose(1, kernel_size=5, strides=2, padding='same',
                                 activation='tanh')
])

# Get the discriminator
discriminator = keras.models.Sequential([
    keras.layers.Conv2D(64, kernel_size=5, strides=2, padding='same',
                        activation=keras.layers.LeakyReLU(0.2),
                        input_shape=[28, 28, 1]),
    keras.layers.Dropout(0.4),
    keras.layers.Conv2D(128, kernel_size=5, strides=2, padding='same',
                        activation=keras.layers.LeakyReLU(0.2)),
    keras.layers.Dropout(0.4),
    keras.layers.Flatten(),
    keras.layers.Dense(1, activation='sigmoid')
])

# Get gan
gan = keras.models.Sequential([generator, discriminator])

# Get the summary
gan.summary()
```

## Compiling the model

```python
# Compile the discriminator
discriminator.compile(loss='binary_crossentropy', optimizer='rmsprop')
# Freeze the discriminator
discriminator.trainable = False

# Compile gan
gan.compile(loss='binary_crossentropy', optimizer='rmsprop')
```

## Training and saving the model

```python
# Get the generator and discriminator
generator, discriminator = gan.layers

# For each epoch
for _ in range(100):
    # For each batch
    for data_batch in data.take(batch_size):
        # Get the image and label
        real_images, labels = data_batch

        # Get the noise
        noise = tf.random.normal(shape=[batch_size, coding_size])

        # Get the fake images
        fake_images = generator(noise)

        # Cominbe the fake and real images
        fake_real_images = tf.concat([fake_images, real_images], axis=0)

        # Get the target
        y = tf.constant([[0.]] * batch_size + [[1.]] * batch_size)

        # Unfreeze the discriminator
        discriminator.trainable = True

        # Train the discriminator
        discriminator.train_on_batch(fake_real_images, y)

        # Get the noise
        noise = tf.random.normal(shape=[batch_size, coding_size])

        # Get the target
        y = tf.constant([[1.]] * batch_size)

        # Freeze the discriminator
        discriminator.trainable = False

        # Train the generator
        gan.train_on_batch(noise, y)

# Save the model
gan.save(abspath + '/model/model.h5')
```

# Testing

## Loading the saved model

```python
# Load the model
model = keras.models.load_model(abspath + '/model/model.h5', custom_objects={'LeakyReLU': keras.layers.LeakyReLU})
```
